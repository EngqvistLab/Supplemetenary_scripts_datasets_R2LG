```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
dfpa = pd.read_csv('results/cross_validation_results_PA_35_phenotypes.csv',index_col=0)
dfcnv = pd.read_csv('results/cross_validation_results_CNV_35_phenotypes.csv',index_col=0)
```

#### Check distribution of these traits 

```python
dfph = pd.read_csv('data/phenoMatrix_35ConditionsNormalizedByYPD.tab',sep='\t',index_col=0)
```

```python
plt.figure(figsize=(13,14))
xlim = [np.amin(dfph.values),np.amax(dfph.values)]
ylim = [0,600]
print(xlim)
for i,col in enumerate(dfph.columns):
    plt.subplot(7,5,i+1)
    plt.hist(dfph[col])
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.text(1.,550,'$\sigma=${:.2f}'.format(np.std(dfph[col])),ha='center',va='center')
    if i%5 != 0: plt.yticks([])
    if i<30: plt.xticks([])
    if i ==32: plt.xlabel('Phenome')
    if i == 15: plt.ylabel('Strain Count')
    plt.title(col)
    
plt.tight_layout()
#plt.savefig('../Results/distribution_of_35_traints.pdf')
plt.show()
```

```python
sigma_db_sqs = list()
for col in dfpa.index:
    sigma_db_sqs.append(np.var(dfph[col]))
```

```python
max_se_sqs = list()
for i,sigma_db_sq in enumerate(sigma_db_sqs):
    r2 = dfpa['r2'][i]
    max_se_sqs.append((1-r2)*sigma_db_sq)

maximal_se_sq = np.min(max_se_sqs)
minimal_seq_sq = np.min(sigma_db_sqs)

print(maximal_se_sq,minimal_seq_sq)
```

```python
for i,sigma_db_sq in enumerate(sigma_db_sqs):
    if sigma_db_sqs[i] == np.min(sigma_db_sqs): print(dfpa.index[i])
```

```python
R2_upper = list()
for sigma_db_sq in sigma_db_sqs:
    R2_upper.append(1-maximal_se_sq/sigma_db_sq)
```

plot r2

```python
plt.figure(figsize=(15,3))
x = np.arange(0,3*len(dfpa.index),3)
plt.bar(x,dfpa['r2'],yerr=dfpa['r2_std'],label='PA')
x+=1
plt.bar(x,dfcnv['r2'],yerr=dfcnv['r2_std'],label='CNV')
plt.xticks(x-0.5,dfpa.index,rotation='90')
plt.scatter(x-0.5,R2_upper,marker='_',color='g',s=200,label='minimal '+ r'$\langle R^2\rangle_{LG}$')
for i,pos in enumerate(x):
    plt.arrow(pos-0.5,R2_upper[i],0,0.05,head_width=0.5,head_length=0.02,color='g')
    print(dfpa.index[i],R2_upper[i])
plt.legend(loc='lower right',ncol=3)
plt.ylabel('$R^2$ from 5-fold cross validation')
#plt.savefig('../Results/cross_validation_results_35_phenotypes.pdf',bbox_inches='tight')
plt.ylim([-0.5,1.1])
plt.xlim([-1.5,len(x)*3])
plt.show()
```
